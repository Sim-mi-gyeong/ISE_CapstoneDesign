```python
import folium
center=[37.480255,126.999028]
m=folium.Map(location=center,zoom_start=11)

for i in range(len(random_place)):
    folium.Marker([random_place.iloc[i][3],random_place.iloc[i][4]],tooltip=random_place.iloc[i][2]).add_to(m)
m
```
