```python
import pandas as pd
from geopy.geocoders import Nominatim

import folium
from haversine import haversine
from math import sin, cos, sqrt, atan2, radians
```

```python
reusable = pd.read_excel('data/다회용컵 시범사업 위치.xlsx')
reusable
```

```python
address = reusable['도로명주소']
address
```

```python
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[:3])
address
```

```python
geo_local = Nominatim(user_agent='South Korea')

def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y
```

```python
latitude = []
longitude =[]

for i in address:
    print(geocoding(i)[0], geocoding(i)[1])
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
reusable['위도'] = latitude
reusable['경도'] = longitude
```

```python
reusable
```

```python
seoul_map = folium.Map(location = [37.541, 126.986]
#                       , tiles = 'Stamen Terrain'
                      , zoom_start = 12)
seoul_map.add_child(folium.LatLngPopup())
```

```python
for i in range(len(reusable)):
    marker = folium.Marker([ reusable.loc[i]['위도'], reusable.loc[i]['경도'] ]
                         , icon = folium.Icon(color='blue')
                         , popup = reusable.loc[i]['도로명주소'])
    marker.add_to(seoul_map)
    
seoul_map
```

```python
data_dic = {}

for i in reusable.index:
  data_dic[reusable.loc[i, '위치']] = (reusable.loc[i, '위도'], reusable.loc[i, '경도'])

data_dic
```

```python
data_list = []

for i in data_dic.values():
  data_list.append(i)  

data_list
```

```python
list_num = len(data_list)
list_num
```

```python
sum = 0
tmpAvgList = []
for x in range(list_num):
    tmpSum = 0
    rowList = []
    for y in range(list_num):
        dist = int(haversine(data_list[x], data_list[y], unit = 'm'))
        tmpSum += dist
        rowList.append(dist)
        print('dist : ', dist)
    tmpAvg = tmpSum / list_num
    tmpAvgList.append(tmpAvg)
```

```python
tmpAvgList
```

```python
sum = 0
for i in tmpAvgList:
    sum += int(i)

print('평균 : ', sum / list_num)
```
