```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
import folium
from folium.plugins import MarkerCluster
from haversine import haversine
from math import sin, cos, sqrt, atan2, radians
```

```python
data = pd.read_csv('data/서초구_위치_전체.csv', encoding = 'CP949')
data
```

```python
# [위도, 경도] 기준으로 중복 제거

data = data.drop_duplicates(['경도', '경도'])
data
```

```python
sub_data_dic = {}

for i in data.index:
#     sub_data_dic[data.loc[i, '분류']] = ( data.loc[i, '위도'], data.loc[i, '경도'] )    # 8 개
    sub_data_dic[( data.loc[i, '위도'], data.loc[i, '경도'] )] = data.loc[i, '분류']
len(sub_data_dic)
```

```python
# {위치 : (위도, 경도), 위치: (위도, 경도), ... }
data_dic = {}

for i in data.index:
    data_dic[i] = ( data.loc[i, '위도'], data.loc[i, '경도'] )
    
data_list = []

for value in data_dic.values():
    data_list.append(value)

listNum = len(data_list)
```

```python
listNum
```

```python
print(data_list[0])
```

```python
x = 0
y = 0
cnt = 0
cnt_del = 0
R = 3000
# print(data_list[0])

while x != listNum - 1:
  y += 1
  try:
    distance = int(haversine(data_list[x], data_list[y], unit = 'm'))
  except:
      break
#   print(x,y)
  if y >= listNum - 1:
    x += 1
    y = x + 1
    # print(x,len(listNum))
    continue

  if distance < R:
    print(x, y, distance)

    del data_list[y]
    y -= 1
    cnt_del += 1    # 지운 데이터 개수
    
  else:
    cnt += 1

print(cnt_del)

# print('-' * 20)

# 검증
# x=0
# y=0
# while x!=len(data_list)-1:
#     y+=1
#     try:
#         distance = int(haversine(data_list[x], data_list[y], unit = 'm'))
#     except:
#         break
#     print(distance)
#     if y >= len(data_list)-1:
#         x += 1
#         y = x+1
#         # print(x,len(data_list))
        
print("SetCovering 후의 남은 데이터 수", len(data_list))

print('-' * 20)

print("후보군 선정 위치 : ", data_list)

# child_data = pd.read_csv('data/은평구_어린이보호구역.csv', encoding = 'CP949')
# print(child_data)

pd_data_list = pd.DataFrame(data_list, columns = [ '위도', '경도' ])
```
