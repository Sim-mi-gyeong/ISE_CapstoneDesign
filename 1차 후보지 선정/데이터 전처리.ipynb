```python
import pandas as pd
```

```python
df = pd.read_csv('서초구_호텔.csv')
df
```

```python
address = df['도로명주소']
address
```

```python
for i in range(len(address)):
    a = address[i].split(' ')
    address[i] = " ".join(a[0:4])
address
```

```python
####### 도로명주소 위도 경도 값으로 바꿔주기 ########
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')
# 위도, 경도 반환하는 함수
def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y
```

```python
latitude = [] # 위도
longitude =[] # 경도

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
address_df = pd.DataFrame({'호텔명':df['사업장명'],'도로명 주소':df['도로명주소'],'위도':latitude,'경도':longitude})
address_df.to_csv('서초구_호텔_위경도.csv')
```
