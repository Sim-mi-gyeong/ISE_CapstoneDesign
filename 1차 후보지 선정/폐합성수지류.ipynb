```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('쓰레기 데이터셋.csv')
df
```

```python
a = list(df['폐합성수지(플라스틱) 발생량(톤/년)'])
b = list(df['생활폐기물 전체 발생량(톤/년)'])
place = list(df['지역'])
for i in range(len(a)):
    b[i] = float(b[i].replace(',',''))
```

```python
plt.plot(place, a)
plt.show()
```

```python
plt.plot(place, b)
plt.show()
```

```python
plt.figure(figsize=(25,18))

fig, ax1 = plt.subplots()
ax1.plot(place, a, color='red', alpha=0.5)

ax2 = ax1.twinx()
ax2.plot(place, b, color='blue', alpha=0.5)

plt.xticks(rotation=45)
plt.show()
```
