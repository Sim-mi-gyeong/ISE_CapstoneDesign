```python
cafeData = cafeData[['상가업소번호', '상호명', '상권업종중분류명', '시군구명', '도로명주소', '경도', '위도']]
cafeData
```

```python
guList = []

for cafe in cafeData['도로명주소']:
    gu = cafe.split()
    guList.append(gu[1])
```
