```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
data = pd.read_excel('data/서초구_대조군_75개.xlsx')
data
```

```python
seoul_map = folium.Map(location = [37.541, 126.986]
#                       , tiles = 'Stamen Terrain'
                      , zoom_start = 12)
seoul_map.add_child(folium.LatLngPopup())
```

```python
for i in range(len(data)):
    marker = folium.Marker([ data.loc[i]['위도'], data.loc[i]['경도'] ]
                         , icon = folium.Icon(color='blue')
                         , popup = data.loc[i]['주소'])
    marker.add_to(seoul_map)
    
seoul_map
```

```python
seoul_map.save('결과/서초구_대조군_75개_시각화.html')
```
